```python
def recommend_flats(data, location=None, bhk_type=None, min_price=None, max_price=None, min_area=None, max_area=None, top_n=5):
    
    filtered_data = data.copy()
    if location:
        filtered_data = filtered_data[filtered_data['location1'] == location]
    if bhk_type:    
        filtered_data = filtered_data[filtered_data['flat_type'] == bhk_type]

    if min_price is not None:
        filtered_data = filtered_data[filtered_data['flat_price'] >= min_price]
    if max_price is not None:
        filtered_data = filtered_data[filtered_data['flat_price'] <= max_price]

    if min_area is not None:
        filtered_data = filtered_data[filtered_data['buildupArea_sqft'] >= min_area]
    if max_area is not None:
        filtered_data = filtered_data[filtered_data['buildupArea_sqft'] <= max_area]

    recommended_flats = filtered_data.sort_values(by='flat_price').head(top_n)
    return recommended_flats
```

```python
recommend_flats(data=data,bhk_type="2 BHK Flat",min_price=0.80,max_price=2.0,min_area=500,max_area=1200,top_n=5)
```
