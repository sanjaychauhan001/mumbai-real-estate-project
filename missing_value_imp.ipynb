```python
X = df.drop(columns=['price'],axis=1)
y = df['price']
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier 
```

```python
le = LabelEncoder()
for i in X.columns:
    if X[i].dtype == 'O':
        X[i] = le.fit_transform(X[i])
```

```python
for_pred = X[X['balcony'].isna()]
```

```python
for_train = X[~X['balcony'].isna()]
```

```python
x_train = for_train.drop(columns=['balcony'], axis=1)
y_train = for_train['balcony']
```

```python
x_test = for_pred.drop(columns=['balcony'], axis=1)
y_test = for_pred['balcony']
```

```python
from sklearn.metrics import accuracy_score
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred = knn.predict(x_train)
    print(i, accuracy_score(y_true=y_train,y_pred=pred))
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
```

```python
p = knn.predict(x_test).tolist()
```
