```python
def age_of_propert(value):
    if value <= 10:
        return 'recent construction'
    elif value >10 and value <=20:
        return 'modern property'
    elif value >20 and value <=50:
        return 'mid age property'
    else:
        return 'old propert'
```

```python
df['age_of_property'] = df['age_of_property'].apply(age_of_propert)
```

```python
df['which_floor'].describe()
```

```python
def on_which_floor(value):
    if value <= 20:
        return 'lower'
    elif value >20 and value <=50:
        return 'middle'
    else:
        return 'higher'
```

```python
df['which_floor'] = df['which_floor'].apply(on_which_floor)
```

```python
df.to_csv('cleaned_level3.csv',index=False)
```
